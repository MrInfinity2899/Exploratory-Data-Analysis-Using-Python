```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv(r"C:\Users\USER\Desktop\Mini Project\imdb_movies_2024_list.csv")

# Clean columns
df['Genre'] = df['Genre'].apply(lambda x: str(x).split(',')[0] if pd.notnull(x) else x)
df['Duration'] = df['Duration'].astype(str).str.extract(r'(\d+)').astype(float)
df['Votes'] = df['Votes'].astype(str).str.replace(",", "").str.extract(r'(\d+)').astype(float)

# Drop missing values in key columns
df = df.dropna(subset=['Rating', 'Votes', 'Genre', 'Duration'])

# ------------------ 1. Top 10 Movies by Rating & Votes ------------------
top_movies = df.sort_values(['Rating', 'Votes'], ascending=[False, False]).head(10)
print("\n Top 10 Movies by Rating and Votes:")
print(top_movies[['Title', 'Rating', 'Votes']])

# ------------------ 2. Genre Distribution ------------------
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 5))
genre_counts.plot(kind='bar', color='skyblue')
plt.title("Genre Distribution")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()

# ------------------ 3. Average Duration by Genre ------------------
avg_duration = df.groupby('Genre')['Duration'].mean().sort_values()
plt.figure(figsize=(8, 6))
avg_duration.plot(kind='barh', color='lightgreen')
plt.title("Average Duration by Genre")
plt.xlabel("Duration (minutes)")
plt.tight_l_
```

```python
avg_votes = df.groupby('Genre')['Votes'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
avg_votes.plot(kind='bar', color='coral')
plt.title("Average Votes by Genre")
plt.ylabel("Average Votes")
plt.tight_layout()
plt.show()

# ------------------ 5. Rating Distribution ------------------
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title("Rating Distribution (Histogram)")

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Rating'])
plt.title("Rating Distribution (Boxplot)")
plt.tight_layout()
plt.show()

# ------------------ 6. Top-Rated Movie per Genre ------------------
top_per_genre = df.loc[df.groupby('Genre')['Rating'].idxmax()]
print("\n🏆 Top-Rated Movie per Genre:")
print(top_per_genre[['Genre', 'Title', 'Rating']].reset_index(drop=True))

# ------------------ 7. Most Popular Genres by Voting ------------------
votes_by_genre = df.groupby('Genre')['Votes'].sum().sort_values(ascending=False)
plt.figure(figsize=(6, 6))
votes_by_genre.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Most Popular Genres by Voting")
plt.ylabel("")
plt.tight_layout()
plt.show()

# ------------------ 8. Shortest and Longest Movies ------------------
shortest = df.loc[df['Duration'].idxmin()]
longest = df.loc[df['Duration'].idxmax()]
print("\n🎬 Shortest Movie:", shortest['Title'], f"({shortest['Duration']} min)")
print("🎥 Longest Movie:", longest['Title'], f"({longest['Duration']} min)")

# ------------------ 9. Heatmap: Ratings by Genre ------------------
heat_data = df.groupby('Genre')['Rating'].mean().to_frame().T
plt.figure(figsize=(10, 2))
sns.heatmap(heat_data, annot=True, cmap="YlGnBu")
plt.title("Average Ratings by Genre (Heatmap)")
plt.tight_layout()
plt.show()

# ------------------ 10. Correlation: Rating vs Votes ------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Votes', y='Rating', hue='Genre', palette='tab10')
plt.xscale('log')
plt.title("Correlation: Rating vs Votes (Log Scale)")
plt.tight_layout()
plt.show()
```
